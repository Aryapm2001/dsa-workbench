```python
acct = {
    "name": "Jimin",
    "acc_no": 13485263816,
    "bal": 500000.0,   # changed to 'bal'
}
acct
```

```python
menu = ["1. Deposit", "2. Withdraw", "3. Check Balance", "4. Exit"]
def show_menu():
    print("Services")
    for ch in menu:
        print(ch)
def deposit(amt):
    acct["bal"] += amt
    print("Transaction successful:", amt)
def withdraw(amt):
    if amt <= acct["bal"]:
        acct["bal"] -= amt
        print("Transaction successful:", amt)
    else:
        print("Insufficient balance")
def check_balance():
    print("Available balance:", acct["bal"])
```

```python
while True:
    show_menu()  
    choice = input("Enter your choice: ")
    if choice == '1':
        amt = int(input("Enter amount to deposit: "))
        deposit(amt)
    elif choice == '2':
        amt = int(input("Enter amount to withdraw: "))
        withdraw(amt)
    elif choice == '3':
        check_balance()
    elif choice == '4':
        break
    else:
        print("Incorrect choice")
```
