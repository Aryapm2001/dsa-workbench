```python
def avg_temp(temp_list):
    return sum(temp_list) / len(temp_list)
def avg_humidity(hum_list):
    return sum(hum_list) / len(hum_list)
def weather_stat(avg_temp, avg_humidity):
    if avg_humidity > 80:
        return "Rainy"
    elif avg_temp > 75:
        return "Sunny"
    elif avg_temp < 40:
        return "Foggy"
    else:
        return "Cloudy"
```

```python
avg_t = avg_temp(temp)
avg_h = avg_humidity(humidity)
status = weather_stat(avg_t, avg_h)
print("Average Temperature:", round(avg_t))
print("Average Humidity:", round(avg_h))
print("Overall Weather Status:", status)
```
